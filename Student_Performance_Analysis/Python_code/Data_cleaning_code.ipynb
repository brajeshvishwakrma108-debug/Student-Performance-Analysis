```python
df=df.drop("Lunch_Type",axis=1)
df
```

```python
df["Total_marks"] = df["Math_Score"] + df["Reading_Score"] + df["Writing_Score"]
df.head()
```

```python
df["Average"] = df["Total_marks"] / 3
df.head()
```

```python
def calculate_grade(avg):
    if avg >= 85:
        return "A"
    elif avg >= 70:
        return "B"
    elif avg >= 50:
        return "C"
    else:
        return "D"

df["Grade"] = df["Average"].apply(calculate_grade)
df.head()
```

```python
df=df.drop("Ethnicity_Group",axis=1)
df
```
